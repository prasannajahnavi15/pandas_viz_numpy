```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
univ=pd.read_csv('Universities.csv')
```

```python
univ
```

```python
plt.hist(univ['SAT'])
plt.xlabel('SAT')
plt.ylabel('count')
plt.title('Histogram of SAT')
plt.show()
```

```python
univ.shape
```

```python
sn.histplot(univ['Accept'])
plt.show()
```

```python
sn.histplot(univ['SAT'])
plt.show()
```

```python
import warnings
warnings.filterwarnings('ignore')
sn.distplot(univ['Accept'])
plt.show()
```

```python
sn.distplot(univ['SAT'])
plt.show()
```

```python
plt.plot(univ['Accept'],univ['SAT'],'ko')
plt.xlabel('Accept')
plt.ylabel('Top10')
plt.title('Scatterplot of Accept Vs Top10')
plt.show()
help(plt.plot)
```

```python
plt.plot(univ['SAT'],univ['Top10'],'ko')
```

```python
sn.pairplot(univ)
```

```python
sn.pairplot(univ.iloc[:,0:2])
```

```python
plt.boxplot(univ['Accept'])
plt.xlabel('Accept')
plt.ylabel('values')
plt.title('Box Plot of Accept')
plt.show()
```

```python
plt.boxplot(univ['Accept'],vert=False)
plt.xlabel('Accept')
plt.ylabel('values')
plt.title('Box Plot of Accept')
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.boxplot(univ['Accept'],vert=False)
plt.xlabel('Accept')
plt.ylabel('values')
plt.title('Box Plot of Accept')
plt.show()
```
